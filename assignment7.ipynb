```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(102316013)

data = np.random.randint(1000, 5001, size=(12, 4))
columns = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df = pd.DataFrame(data, columns=columns, index=months)

df.head()
df.describe()

total_sales_category = df.sum()
total_sales_month = df.sum(axis=1)

growth_rates = df.pct_change().fillna(0) * 100

df['Total Sales'] = df.sum(axis=1)
df['Growth Rate'] = df['Total Sales'].pct_change().fillna(0) * 100

if 102316013 % 2 == 0:
    df['Electronics'] = df['Electronics'] * 0.9
else:
    df['Clothing'] = df['Clothing'] * 0.85

plt.figure(figsize=(10,6))
for col in columns:
    plt.plot(df.index, df[col], label=col)
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(data=df[columns])
plt.title('Sales Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()

array = np.array([[1, -2, 3], [-4, 5, -6]])
abs_array = np.abs(array)
percentiles_25 = np.percentile(array.flatten(), 25)
percentiles_50 = np.percentile(array.flatten(), 50)
percentiles_75 = np.percentile(array.flatten(), 75)
mean_flatten = np.mean(array.flatten())
median_flatten = np.median(array.flatten())
std_flatten = np.std(array.flatten())
mean_columns = np.mean(array, axis=0)
mean_rows = np.mean(array, axis=1)
median_columns = np.median(array, axis=0)
median_rows = np.median(array, axis=1)
std_columns = np.std(array, axis=0)
std_rows = np.std(array, axis=1)

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
floor_values = np.floor(a)
ceil_values = np.ceil(a)
trunc_values = np.trunc(a)
rounded_values = np.round(a)

def swap_list_elements(lst, i, j):
    temp = lst[i]
    lst[i] = lst[j]
    lst[j] = temp
    return lst

lst = [1, 2, 3, 4]
swap_list_elements(lst, 1, 3)

def swap_set_elements(s, elem1, elem2):
    lst = list(s)
    idx1, idx2 = lst.index(elem1), lst.index(elem2)
    lst[idx1], lst[idx2] = lst[idx2], lst[idx1]
    return set(lst)

s = {1, 2, 3, 4}
swap_set_elements(s, 2, 4)
```
